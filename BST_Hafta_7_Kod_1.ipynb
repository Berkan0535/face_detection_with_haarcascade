```python
import cv2
import numpy as np
```

```python
my_bgr_img = cv2.imread("./bonus.jpg")
my_gray_img = cv2.imread("./bonus.jpg",0)
```

```python
ret, thresh = cv2.threshold(my_gray_img, 127, 255, cv2.THRESH_BINARY)

contour, hierachy = cv2.findContours(thresh, cv2.RETR_EXTERNAL,
                                    cv2.CHAIN_APPROX_SIMPLE)

cv2.drawContours(my_bgr_img, contour, -1, (0,255,0),2)

cv2.imshow("CONTOURED IMAGE",my_bgr_img)
cv2.imshow("THRESH IMAGE",thresh)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
ret, thresh = cv2.threshold(my_gray_img, 127, 255, cv2.THRESH_BINARY_INV
                           + cv2.THRESH_OTSU)

thresh = cv2.morphologyEx(thresh, cv2.MORPH_CLOSE, 
                          np.ones((3,3),dtype=np.uint8), iterations=2)

contour, hierachy = cv2.findContours(thresh, cv2.RETR_EXTERNAL,
                                    cv2.CHAIN_APPROX_SIMPLE)

cv2.drawContours(my_bgr_img, contour, -1, (0,255,0),2)

cv2.imshow("CONTOURED IMAGE",my_bgr_img)
cv2.imshow("THRESH IMAGE",thresh)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
