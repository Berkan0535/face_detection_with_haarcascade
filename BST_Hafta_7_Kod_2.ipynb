```python
import cv2
import numpy as np
```

```python
# Yüz tanıma için daha önceden eğitilmiş bir modelin ağırlıklarını
# indiriyoruz (internet bağlantısı önemli)

model_pathh = "haarcascade_frontalface_default.xml"

face_cascades = cv2.CascadeClassifier(cv2.data.haarcascades + model_pathh)
```

```python
my_bgr_img = cv2.imread("./deneme.jpg")
my_gray_img = cv2.imread("./deneme.jpg",0)
```

```python
# İndirdiğimiz ağırlıklardan meydana gelen model, bir yüzü seçerken (tanımlarken)
# hangi parametrelere göre o yüzü seçecek/tespit edecek? Aşağıdaki kod
# satırıyla bu işlemi yerine getiriyoruz

faces = face_cascades.detectMultiScale(
    my_gray_img, # her zaman gri seviye görüntü işlenir
    scaleFactor = 1.1, # yakınlaştırma oranı (%10 yakınlaştırır)
    minNeighbors = 5, # yanlış pozitifleri elemek için danışacağı komşu sayısı
    minSize = (5,5)
)
```

```python
# Yüzleri seçerken, onları bir dikdörtgen içine alıp, o şekilde ekrana
# çizdirmek için aşağıdaki satırları oluşturuyoruz

for (x,y,w,h) in faces:
    cv2.rectangle(my_bgr_img,(x,y),(x+w,y+h),(0,255,0),2)

cv2.imshow("EFSANE FENERBAHCE KADROSU",my_bgr_img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
